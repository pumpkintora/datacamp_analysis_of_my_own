```python
-- Select the names, continents, and areas of countries
-- Note that the query may take a long time to run if you remove the LIMIT statement
SELECT
	countries.name AS country_name,
	continents.name AS continent_name,
	area
FROM countries
	INNER JOIN regions USING(region_id)
	INNER JOIN continents USING(continent_id)
ORDER BY country_name
LIMIT 5
```

```python
-- countries with no official language
SELECT 
    countries.name AS country_name,
    SUM(country_languages.official) AS num_official_language
FROM country_languages
    INNER JOIN countries USING(country_id)
GROUP BY country_name
HAVING num_official_language = 0;
```

```python
-- ranking of gdp / country area in 2018
SELECT 
    c.name AS country_name,
    c.area,
    cs.gdp,
    (cs.gdp / c.area) AS gdp_per_area
FROM 
    nation.countries c
JOIN 
    nation.country_stats cs ON c.country_id = cs.country_id
WHERE cs.year = 2018
ORDER BY 
    gdp_per_area DESC;
```

```python
-- average gdp of every countries
SELECT c.name, c.area, c.country_id,
(SELECT AVG(cs.gdp) FROM nation.country_stats AS cs WHERE cs.country_id = c.country_id) AS avg_gdp
FROM nation.countries AS c
ORDER BY avg_gdp DESC;
```

```python
-- average gdp of most recent 5 years
WITH recent_5_years AS (
	SELECT DISTINCT year
        FROM country_stats
        ORDER BY year DESC
        LIMIT 5
),
cs_recent_5_years AS (
	SELECT * FROM country_stats
		WHERE country_stats.year IN 
		(SELECT year FROM recent_5_years)
)

SELECT c.name, c.region_id, AVG(cs.gdp) AS average_gdp_last_5_years
FROM countries AS c
LEFT JOIN cs_recent_5_years AS cs USING(country_id)
GROUP BY c.country_id
ORDER BY average_gdp_last_5_years DESC;
```

```python
-- country with highest gdp in every region (year 2010)
-- SELECT * FROM regions;
SELECT c.name AS country, c.country_id, cs.gdp AS highest_gdp, r.name AS region_name
FROM country_stats AS cs
LEFT JOIN countries AS c ON cs.country_id = c.country_id
LEFT JOIN regions AS r ON c.region_id = r.region_id
WHERE cs.year = 2010
AND cs.gdp = (
    SELECT MAX(cs2.gdp)
    FROM country_stats AS cs2
    LEFT JOIN countries AS c2 ON cs2.country_id = c2.country_id
    WHERE cs2.year = 2010 AND c2.region_id = r.region_id
)
ORDER BY highest_gdp DESC;
```
